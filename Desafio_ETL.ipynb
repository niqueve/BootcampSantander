```python
uf = list(dados['UF'])
```

```python
def estados (uf):
  tabela = []
  for n in uf:
    if n in tabela:
      None
    else:
      tabela.append(n)
  return tabela
```

```python
estados_uf = estados(uf) # Lista com todos os estados
```

```python
estados_uf = estados_uf[2:] # retirar Nan
```

```python
estados_uf
```

```python
def total_estacoes (estados,dados):
  tabela = []
  for n in estados:
    total = dados.count(n)
    tabela.append([n,total])
  dataframe = pd.DataFrame(data = tabela, columns = ['Estados','Total de Estações'])
  return dataframe
```

```python
total = total_estacoes(estados_uf,uf)
```

```python
total.to_csv('total_estações.csv') # salva arquivo com total de estações em csv
```
